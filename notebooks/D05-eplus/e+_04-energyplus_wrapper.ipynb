# Energyplus Wrapper, une librairie automatisant le lancement et le traitement des données E+.

Cette librairie a été créé pour rendre plus simple le lancement de nombreuses simulation EPlus.

```python
from path import Path
from energyplus_wrapper import EPlusRunner
```

```python
runner = EPlusRunner('/home/nicolas/.local/share/energy_plus_wrapper/EnergyPlus-8-7-0')
```

```python
idf_file = Path("../../data/eplus_data/in_8-7-0.idf")
epw_file = Path("../../data/eplus_data/in.epw")
```

```python
eplus_sim = runner.run_one(idf=idf_file, epw_file=epw_file)
```

```python
eplus_sim.time_series["eplus"]
```

Elle inclue une gestion de la parallélisation pour le lancement de nombreux runs en parallèle via `joblib` :

```python
from joblib import parallel_backend
```

```python
samples = {f"run_{i:02g}": (idf_file, epw_file) for i in range(4)}
```

```python
samples
```

```python
with parallel_backend("loky"):
    sims = runner.run_many(samples)
```
