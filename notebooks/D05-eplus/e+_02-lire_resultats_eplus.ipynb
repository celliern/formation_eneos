# Lecture des résultats Energy Plus

Comme vu précédement, les EnergyPlus génère beaucoup de fichiers différents. Selon le fichier IDF donnée en entrée, il est possible de récupérer les données CSV de la simulation.

## Les fichiers CSV

Ils sont relativement standard : la date peut être géré automatiquement par Pandas. Il faudra juste faire attention à la présence d'un pied de fichier

```python
out_dir = Path("./output_eplus/")
```

```python
print(*out_dir.files("*.csv"), sep="\n")
```

```python
df = pd.read_csv("./output_eplus/eplusssz.csv",
                 parse_dates=True, index_col=0,
                 skipfooter=2)
df.columns
```

```python
df = pd.read_csv("./output_eplus/epluszsz.csv",
                 parse_dates=True, index_col=0,
                 skipfooter=2)
df.columns
```

## Le rapport, lecture du fichier HTML

EPlus génère également un rapport global, en plus des séries temporelles. Il est possible de le sortir en csv, mais c'est déconseillé : le fichier est *très* mal formaté et contient plusieurs tableaux. Au lieu de bricoler le fichier csv pour le lire, il vaut mieux lire directement le fichier html généré.

```python
from IPython.display import display_html, display
```

```python
with open("output_eplus/eplustbl.htm") as f:
    display_html(f.read(), raw=True)
```

```python
dfs = pd.read_html("output_eplus/eplustbl.htm", index_col=0, header=0)
```

```python
for i, df in enumerate(dfs):
    print(i)
    print(df.index)
```
